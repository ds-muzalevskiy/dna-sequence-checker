```python
import csv
                                
def fastq_to_csv(fastq_path, csv_path):
    with open(fastq_path, 'r') as fastq, open(csv_path, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile, quoting=csv.QUOTE_NONE)
        writer.writerow(['sequence'])  
        
        while True:
            identifier = fastq.readline()
            sequence = fastq.readline()
            plus = fastq.readline()
            quality = fastq.readline()
            
            if not quality:
                break
            
            writer.writerow([sequence.strip()])
```
